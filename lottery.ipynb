```python
import requests
import json 
import pandas as pd 
import csv 
from lxml import etree 
url = "http://datachart.500.com/ssq/history/newinc/history.php?start=00001&end=18135" 
response = requests.get(url) 
response = response.text 
selector = etree.HTML(response)
datas = list()
for i in selector.xpath('//tr[@class="t_tr1"]'): 
    data = list()
    data.append(i.xpath('td/text()')[0])
    data.append(i.xpath('td/text()')[15])
    data.append(i.xpath('td/text()')[1])
    data.append(i.xpath('td/text()')[2])
    data.append(i.xpath('td/text()')[3])
    data.append(i.xpath('td/text()')[4])
    data.append(i.xpath('td/text()')[5])
    data.append(i.xpath('td/text()')[6])
    data.append(i.xpath('td/text()')[7])
    data.append(i.xpath('td/text()')[14].replace(',',''))
    data.append(i.xpath('td/text()')[9].replace(',',''))
    data.append(i.xpath('td/text()')[10].replace(',',''))
    data.append(i.xpath('td/text()')[11].replace(',',''))
    data.append(i.xpath('td/text()')[12].replace(',',''))
    data.append(i.xpath('td/text()')[13].replace(',',''))
    datas.append(data)
df1 = pd.DataFrame(datas, columns=['code', 'date', 'red1','red2','red3','red4','red5','red6', 'blue', 'sales', 'poolmoney', 'pour1', 'money1','pour2','money2'])
df1
```

```python
df = df1['red1'].to_frame().astype(int)
df
```

```python
import numpy as np
seq_len = 100
input_train = []
output_train = []
input_test = []
output_test = []
stock_train = df[0: int(0.9 * len(df))]
stock_test = df[int(0.9 * len(df)):]
for i in range((len(stock_train)//seq_len)*seq_len - seq_len - 1):
    x = np.array(stock_train.iloc[i: i + seq_len, 0])
    y = np.array([stock_train.iloc[i + seq_len + 1, 0]], np.float64)
    input_train.append(x)
    output_train.append(y)
X_train = np.array(input_train)
Y_train = np.array(output_train)

for i in range((len(stock_test)//seq_len)*seq_len - seq_len - 1):
    x = np.array(stock_test.iloc[i: i + seq_len, 0])
    y = np.array([stock_test.iloc[i + seq_len + 1, 0]], np.float64)
    input_test.append(x)
    output_test.append(y)
X_test = np.array(input_test)
Y_test = np.array(output_test)

X_train = X_train / 100
Y_train = Y_train / 100

X_test = X_test / 100
Y_test = Y_test / 100
```

```python
X_train 
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.relu))

model.compile(optimizer="adam", loss="mean_squared_error")

model.fit(X_train, Y_train, epochs=100, verbose = 0)

print(model.evaluate(X_test, Y_test))
```

```python
import numpy as np
seq_len = 100
input_train = []
output_train = []
input_test = []
output_test = []
stock_train = df[0: int(0.9 * len(df))]
stock_test = df[int(0.9 * len(df)):]
for i in range((len(stock_train)//seq_len)*seq_len - seq_len - 1):
    x = np.array(stock_train.iloc[i: i + seq_len, 0])
    y = np.array([stock_train.iloc[i + seq_len + 1, 0]], np.float64)
    input_train.append(x)
    output_train.append(y)
X_train = np.array(input_train)
Y_train = np.array(output_train)

for i in range((len(stock_test)//seq_len)*seq_len - seq_len - 1):
    x = np.array(stock_test.iloc[i: i + seq_len, 0])
    y = np.array([stock_test.iloc[i + seq_len + 1, 0]], np.float64)
    input_test.append(x)
    output_test.append(y)
X_test = np.array(input_test)
Y_test = np.array(output_test)

X_train = X_train.reshape((len(X_train), 100, 1)) / 100
Y_train = Y_train / 100

X_test = X_test.reshape((len(X_train), 100, 1)) / 100
Y_test = Y_test / 100
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(20, input_shape=(100, 1), return_sequences=True))
model.add(tf.keras.layers.LSTM(20))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.relu))

model.compile(optimizer="adam", loss="mean_squared_error")

model.fit(X_train, Y_train, epochs=100,verbose = 0)

print(model.evaluate(X_test, Y_test))
```
