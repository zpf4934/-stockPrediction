```python
import get_prices as hist
import pandas_datareader.data as pdr
import numpy as np
import tensorflow as tf
from preprocessing import DataProcessing
# import pandas_datareader.data as pdr if using the single test below
import fix_yahoo_finance as fix
fix.pdr_override()

start = "2003-01-01"
end = "2018-01-01"

# hist.get_stock_data("BABA", start_date=start, end_date=end)
process = DataProcessing("stock_prices.csv", 0.9)
process.gen_test(10)
process.gen_train(10)

X_train = process.X_train / 200
Y_train = process.Y_train / 200

X_test = process.X_test / 200
Y_test = process.Y_test / 200

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.relu))

model.compile(optimizer="adam", loss="mean_squared_error")

model.fit(X_train, Y_train, epochs=100, verbose = 0)

print(model.evaluate(X_test, Y_test))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# data = pdr.get_data_yahoo("BABA", "2017-12-19", "2018-11-01")
# data.to_csv("predict_prices.csv")
data = pd.read_csv("predict_prices.csv",index_col = 'Date')
stock_data = data["Adj Close"]
errors = []
analyse = []
date = []
for i in range((len(stock_data)//10)*10 - 10 - 1):
    x = np.array(stock_data.iloc[i: i + 10]).reshape((1,10,1)) / 200
    y = np.array(stock_data.iloc[i + 10 + 1]) / 200
    date.append(pd.to_datetime(stock_data.index[i + 10 + 1]))
    predict = model.predict(x)
    while predict == 0:
        predict = model.predict(x)
    analyse.append([predict[0][0]*200,y*200,y*200 - predict[0][0]*200])
    error = (predict - y) * 200
    errors.append(error)
    total_error = np.array(errors)
print(f"Average error = {total_error.mean()}")
# If you want to see the full error list then print the following statement
# print(total_error)
errorlist = [i[0] for i in total_error.tolist()]

DFerror = pd.DataFrame(errorlist,columns = ['poor'])
DFerror.plot(kind = 'line',grid = True,figsize = (15,8))
plt.show()
```

```python
dd = pd.DataFrame(analyse,columns = ['predict','real','D-value'], index = date)
dd[['predict','real']].plot(kind = 'line',grid = True,figsize = (15,8),title = 'BABA')
plt.show()
```
